### Demonstrate the EcoBase interface for SpatialEcology objects

```julia
using CSV, DataFrames, SpatialEcology
amphdata = CSV.read(joinpath(dirname(pathof(SpatialEcology)), "..", "data", "amph_Europe.csv"));
amph = Assemblage(amphdata[4:end],amphdata[1:3], sitecolumns = false)
```

```julia
plot(amph)
```

```julia
occurrences(amph)
```

```julia
placenames(amph)
```

```julia
places(amph)
```

### `getcoords` returns an object with the locations of each site

```julia
cd = getcoords(places(amph))
```

```julia
dump(cd)
```

### That's because the plot methods in SpatialEcology are all intended to create maps.
So here, let's add a new variable to the object and plot it

```julia
using StatsBase
ord = ordinalrank(richness(amph))
addsitestats!(amph, ord, :richnessrank);
```

```julia
plot(:richnessrank, amph)
```

```julia
meta_gamma(amph, 1)
```

```julia
all(asna[:diversity] .≈ richness(amph))
```
